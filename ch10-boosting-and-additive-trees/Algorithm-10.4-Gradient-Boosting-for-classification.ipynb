```python
import altair
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats
import sklearn.model_selection
import sklearn.preprocessing
import sklearn.tree
from tqdm import tqdm

import gb_classification

np.random.seed(123)

%load_ext autoreload
%autoreload 2
```

We'll implement gradient boosting algorithm for classification and then apply it to a toy classification problem.

# Make data

```python
df = gb_classification.make_data(num=500)
```

```python
df.label.value_counts()
```

```python
altair.Chart(df).mark_circle().encode(x='x1:Q', y='x2:Q', color='label:N')
```

```python
model = gb_classification.GradientBoostingClassification(
    learning_rate=1,
    max_depth=5,
    n_estimators=10,
)
```

```python
model.fit(
    df[["x1", "x2"]],
    sklearn.preprocessing.OneHotEncoder().fit_transform(df[["label"]]).todense(),
)
```

```python
df_preds = df.assign(predicted=lambda df: model.predict(df[["x1", "x2"]])).assign(
    match=lambda df: 'truth: ' + df.label.astype(str) + ' predicted: ' + df.predicted.astype(str)
)
```

```python
altair.Chart(df_preds).mark_circle().encode(x='x1:Q', y='x2:Q', color='match:N')
```
